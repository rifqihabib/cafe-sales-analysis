```python
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("cafe_sales_cleaned_final.csv")
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df.head()
```

```python
df['Year'] = df['Transaction Date'].dt.year
df['Month'] = df['Transaction Date'].dt.month
df['Day'] = df['Transaction Date'].dt.day
df['Day of Week'] = df['Transaction Date'].dt.day_name()
```

```python
df['Is Weekend'] = df['Day of Week'].isin(['Saturday', 'Sunday'])
```

```python
def categorize_spending(amount):
    if amount < 10:
        return 'Low'
    elif amount < 25:
        return 'Medium'
    else:
        return 'High'

df['Spending Category'] = df['Total Spent'].apply(categorize_spending)
```

```python
df['Effective Unit Price'] = df['Total Spent'] / df['Quantity']
```

```python
df_encoded = pd.get_dummies(df, columns=['Payment Method', 'Location'], drop_first=True)
df_encoded.head()
```

```python
df.to_csv("cafe_sales_featured.csv", index=False)

from google.colab import files
files.download("cafe_sales_featured.csv")
```
